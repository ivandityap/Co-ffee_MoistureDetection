```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
name = 'TDCNN'
path1 = name+'.csv'
```

```python
data1 = pd.read_csv(path1)
data1.head()
```

```python
x = [i for i in range(1,26)]
accuracy = data1['accuracy']
val_accuracy = data1['val_accuracy']
val_loss = data1['val_loss']
loss = data1['loss']
```

```python
markersize = 6
linewidth = 1.5

mpl.style.use('seaborn')
plt_1 = plt.figure(figsize=(5,4),dpi=300)
plt.plot(x, accuracy,color='blue', marker='o', label='Train Acc', linewidth=linewidth, markersize=markersize)
plt.plot(x, val_accuracy, color='orange', marker='D', label='Val Acc', linewidth=linewidth, markersize=markersize)
plt.plot(x, loss, color='green', marker='s', label='Train Loss', linewidth=linewidth, markersize=markersize)
plt.plot(x, val_loss, color='red', marker='X', label='Val Loss', linewidth=linewidth, markersize=markersize)
plt.axis([0,26,0.38,1.05])
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy/Loss')
plt.title('T-DCNN Training')

plt.show()



```

```python
image_format = 'png' # e.g .png, .png, etc.
image_name = name+'.png'

plt_1.savefig(image_name, format=image_format, dpi=300)
```

```python
import csv

name = 'TDCNN'

lines = []
data = []
with open(name+'.txt') as f:
    lines = f.readlines()

for line in lines[1::2]:
    line = line.split()
    loss = float(line[7]) 
    accuracy = float(line[10])  
    val_loss = float(line[13])  
    val_accuracy = float(line[16])  
    lr = float(line[19])
    data.append([loss,accuracy,val_loss,val_accuracy,lr])

head = ['loss','accuracy','val_loss','val_accuracy','lr'] 

with open(name+'.csv', 'w') as f:
      
    # using csv.writer method from CSV package
    write = csv.writer(f)
    write.writerow(head)
    write.writerows(data)
```
